3. Test del chi-quadro $\chi^2$

Nel seguente esercizio è stato implementato il test del chi-quadro per verificare la bontà del generatore di numeri casuali adottato per i punti precedenti (e che verrà usato anche per le prossime esercitazioni), ossia che generi bene i numeri pseudo-casuali secondo una distribuzione uniforme $[0,1)$.

Per fare ciò sono stati generati casualmente $N = 10^6$ numeri casuali $\sim U([0,1))$ (distribuiti uniformemente tra $[0,1)$) e diviso l'intervallo di generazione in $M = 10^2$ blocchi. Sono stati contati poi i punti $n_i$ appartententi a ciascun intervallo (blocco) e calcolato: $$\chi^2 = \sum_{i=1}^M \frac{\left( n_i - <n_i> \right)^2}{\sigma_{n_i}}\, ,$$
dove $<n_i>$ è il valore atteso del numero di punti per ognuno dei 100 intervalli ed $\sigma_{n_i}$ l'errore associato.

Dal momento che la probabilità che $n_i$ punti appartengano all'intervallo i-esimo segue una distribuzione binomiale ($<n_i> = \frac{n}{M}$ e $\sigma^2= n \frac{1}{M}(1- \frac{1}{M}) \approx  \frac{n}{M}$), si può riscrivere la relazione precedente nella seguente maniera: $$\chi^2 = \sum_{i=1}^M \frac{\left( n_i - n/M \right)^2}{n/M} \, .$$

Il grafico sottostante mostra i valori del $\chi^2$ in funzione di ciascun intervallino $j$. Si osserva come i valori del chi-quadro oscillino attorno al valore atteso che è 100. Dal momento che le fluttuazioni non presentano particolari anomalie e che non si hanno divergenze in determinati punti, si può affermare che il generatore di numeri casuali sia un bun generatore. 

```python
plt.figure(figsize=(10,5))
M, chi= np.loadtxt('Risultati/Risultati01.1.3.out', usecols=(0,1), delimiter=' ', unpack='true')
plt.title('Test del $\chi^2$')
plt.plot(M,chi)
plt.xlabel(r'$j$')
plt.ylabel(r'$\chi^2$')
plt.grid(True)
plt.show()
```

### Exercise 01.2

Lo scopo di questo esercizio è verificare la validità del teorema del limite centrale. 

Per fare ciò ho eteso il generatore di numeri casuali con la possibilità di generare numeri distribuiti secondo una gaussiana a media e deviazione standard fissate, una distribuzione esponenziale $p(x) = \lambda \exp(-\lambda x)$, $x\in [0;+\infty]$ e la distribuzione di Cauchy (o Lorentziana) $p(x)=\frac{1}{\pi}\frac{\Gamma}{(x-\mu)^2+\Gamma^2}$, $x\in [-\infty;+\infty]$. La tecnica che ho adottato per effettuare questa estensione del mio generatore è quella dell'inversione della distribuzione cumulativa.

In seguito ho preparato la mia simulazione. Questa prevede $M = 10^4$ lanci di $N=1, 2, 10, 100$ dadi e la valutazione per ciascun lancio della media $S_N = \frac{1}{N}\sum_{i=1}^N x_i$, dove $x_i$ è una variabile casuale campionata facendo dei lanci standard, esponenziali ($\lambda=1$) o Lorentziani ($\mu=0$ e $\Gamma=1$).

Il teorema del limite centrale dice che se si hanno $N$ variabili casuali statisticamente indipendenti e identicamente distribuite secondo una distribuzione a varianza finita allora si ha che la variabile media $S_N$ è asintoticamente distribuita secondo una gaussiana. Osservando gli istogrammi riportati nelle celle sottostanti si osserva infatti che per i lanci uniformi e esponenziali (distribuiti secondo densità di probabilità a varianza finita) al crescere di $N$ (e quindi del numero di dadi per ciascun lancio) la distribuzione degli $S_N$ approssima sempre meglio quello di una gaussiana (lo confermano i fit eseguiti sui lanci con 100 dadi).

Al contrario, per i lanci Lorentziani non si ha questo tipo di convergenza dal momento che, essendo la distribuzione di Cauchy a varianza infinita, non è applicabile il teorema del limite centrale. Tuttavia, essendo questa distribuzione invariante sotto convoluzione e quindi stabile, essa soddisfa il teorema di Levy e Khintchine. Ciò lo conferma gli istogrammi eseguiti per lanci lorentziani: essendo stabile, la distribuzione di Cauchy è una distribuzione limite e quindi la distribuzione degli $S_N$ generati lorentzianamente tenderà a una distribuzione di Cauchy.

### Exercise 01.3


Lo scopo di questo esercizio è quello di calcolare il valore di $\pi$ mediante l'esperimento di Buffon.

L'esperimento di Buffon prevede che un ago di lunghezza $L$ venga lanciato casualmente su un piano diviso da linee orizzontali poste a una distanza $d>L$ l'una dall'altra. La probabilità che l'ago tocchi una delle righe è $P = 2L/\pi d$. Perciò questo esperimento permette una stima del $\pi$ mediante la formula: 

$$\pi = \frac{2L}{Pd} = \lim_{N_{thr} \to \infty}\frac{2LN_{thr}}{N_{hit}d} \, ,$$ 

dove $N_{thr}$ è il numero di lanci di aghi effettuato e $N_{hit}$ è il numero di volte in cui l'ago ha toccato una riga sul piano.

Per simulare questo esperimento ho utilizzato come parametri $d = 6$ e $L = 4$. Dal momento che il sistema possiede una simmetria per piani paralleli, è equivalente che io mi soffermi nel considerare un sistema di riferimento con $-d/2<x<d/2$, $-\infty<y<+\infty$) e avente in $x = 0$ una delle righe. Di conseguenza per simulare il lancio di un ago ho semplicemente estratto casualmente il centro dell'ago nel range $[-d/2,d/2]$ e inoltre la sua inclinazione una volta atterrato sul piano e quindi il seno dell'angolo che l'ago descrive rispetto alla riga. La generazione casuale del seno dell'angolo (in modo da non utilizzare il $\pi$ nella procedura) l'ho eseguita lanciando punti $(x,y)$ sulla semicirconferenza positiva di raggio 1 e valutandone il seno. Una volta ottenuto il centro dell'ago e la sua inclinazione ho controllato che gli estremi dell'ago non intersecassero $x=0$ (ossia la riga); in caso contrario incrementavo il contatore di $N_{hit}$.

La mia simulazione prevede un numero $M = 10^4$ di esperimenti di Buffon in ciascuno dei quali lancio $N_{thr} = 3\times 10^4$ aghi sul piano. Infine da ogni esperimento estraggo la stima di $\pi$ e per i $M$ valori ottenuti ho eseguito il *data-blocking* con un numero di blocchi $N_{block} = 100$. 

Nel grafico sottostante persento i risultati della simulazione con annessi errori al variare di $N_{block}$. Si osserva come i valori ottenuti con la tecnica Monte Carlo convergano bene al valore vero di $\pi$.


```python
plt.figure(figsize=(10,5))
N, r, error = np.loadtxt('Risultati/Risultati01.3.out', usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(N,r,yerr=error,label = 'Monte Carlo')
plt.axhline(y=np.pi, color = 'red',linestyle='-',label = 'Valore vero')
plt.title('Calcolo del valore di $\pi$ mediante esperimento di Buffon')
plt.xlabel('$N_{block}$')
plt.ylabel('$\pi$')
plt.legend()
plt.grid(True)
plt.show()
```
