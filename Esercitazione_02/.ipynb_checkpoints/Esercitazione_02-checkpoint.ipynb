### Exercise 02.1


1. Calcolo dell'integrale  

$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$

adottando il campionamento di una distribuzione uniforme.

Per questa prima parte dell'esercizio non ho fatto altro che applicare il metodo della media per il calcolo di $I$. Di conseguenza ho generato casualmente $K$ numeri $x_i$ distribuiti uniformemente tra $[0,1)$ su cui valutare l'integranda $f$ e come stima dell'integrale ho adottato la relazione:

$$I = (b-a)\frac{1}{K} \sum_{i=1}^K f(x_i) \, ,$$ 

dove $a$ e $b$ sono gli estremi di integrazione. 

Dunque per ottenere una buona stima di $I$ ho calcolato $M = 10^4$ integrali e ho applicato il metodo del *data blocking* con un numero di blocchi $N_{block} = 100$. Si riporta nella cella sottostante il grafico  al variare di $M$ del valore di $I$ ottenuto con le medie progressive accompagnato dai corrispondenti errori. Si osserva come il risultato dell'integrale converga alla soluzione analitica all'aumentare di $M$; questo comportamente è accompagnato da una diminuzione progressiva dell'errore come garantito dal teorema del limite centrale (TLC). Infatti in questo caso TLC è valido dal momento che ciascun integrale corrisponde a una somma di variabili indipendenti e identicamente distribuite secondo una distribuzione uniforme con varianza finita.

```python
plt.figure(figsize=(10,5))
N, r, error = np.loadtxt('Risultati/Exercise_02.1/Risultati02.1.out', usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(N,r,yerr=error,label = 'Monte Carlo')
plt.axhline(y=1, color = 'red',linestyle='-',label = 'Soluzione analitica')
plt.xlabel('$M$')
plt.ylabel('$I$')
plt.legend()
plt.title('Calcolo integrale $I$ con campionamento distribuzione uniforme [0,1)')
plt.grid(True)
plt.show()
```

2.  Calcolo dell'integrale  

$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$

adottando la tecnica dell'*importance sampling*.

Per applicare il metodo dell'*importance sampling* ho cercato una densità di probabilità che fosse il più simile all'integranda ($\frac{\pi}{2}\cos(\pi x/2)$). Ho deciso di adottare per una retta passante per $\pi/2$ in $x=0$ con pendenza negativa: 

$$z(x) = \frac{\pi}{2}(1-x) \, .$$

Dunque per adottare l'*importance sampling* ho dovuto normalizzare $z(x)$ in modo da ottenere una nuova distribuzione di probabilità con cui valutare l'integrale:

$$p(x) = 2(1-x) \, .$$ 

In seguito per poter generare numeri casuali secondo la $p(x)$ ho utilizzato il metodo di inversione della distribuzione cumulativa. Dunque ho calcolato la cumulativa di $p(x)$: 

$$F(x) = 2x-x^2 \, .$$

Infine ho invertito la $F(x)$ ottenendo quindi la formula per campionare la $p(x)$: $x = 1+\sqrt{1-y}$, dove $y\sim U\left(\left[0,1\right)\right)$. 

Dunque ho applicato la formula precedente per il calcolo dell'integrale esposta per il punto 1; unica differenza è che gli $x_i$ sono distribuiti secondo la $p(x)$, mentre l'integranda diventa:

$$f(x) = \frac{\pi \cos(\pi x/2)}{4(1-x)} \, .$$

Per la stima di $I$, ho calcolato $M = 10^4$ integrali e ho applicato il metodo del *data blocking* con un numero di blocchi $N_{block} = 100$. Si riporta nella cella sottostante il grafico  al variare di $M$ del valore di $I$ ottenuto con le medie progressive accompagnato dai corrispondenti errori. Le considerazioni effettuate nel punto 1 sono applicabili anche a questo risultato. 

Tuttavia è bene osservare come la stima di $I$ tramite *importance sampling* risulti più precisa rispetto a quella precedente. Infatti in questo caso le barre d'errore sono decisamente più piccole (si scenda di due celle per la verifica numerica).

```python
plt.figure(figsize=(10,5))
N2, r2, error2 = np.loadtxt('Risultati/Exercise_02.1/Risultati02.2.out', usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(N2,r2,yerr=error2,label ='Monte Carlo')
plt.axhline(y=1, color = 'red',linestyle='-',label = 'Soluzione analitica')
plt.xlabel('$M$')
plt.ylabel('$I$')
plt.title('Calcolo integrale $I$ con importance sampling')
plt.legend()
plt.grid(True)
plt.show()
```

### Exercise 02.2


Per eseguire il seguente esercizio è stata creata una classe ad hoc (file <span style="color:red">emcee.cpp</span> e <span style="color:red">emcee.h</span>) che fosse in grado di simulare sia *random walk* discreto su un reticolo cubico con passo reticolare $a$ fissato sia in uno spazio tridimesionale continuo con passo fissato $a$. Entrambi i *random walk* sono stati computati in modo che partissero dall'origine delle coordinate $\vec{r} = \vec{0}$. In seguito ciascun passo viene generato come segue:

- ***Random walk* discreto**: viene estratto un numero casuale intero $i ={0,1,2,3,4,5}$, ciascuno corrispondente a un preciso spostamento nel reticolo cubico (ad esempio se i = 1 allora si sposta di uno indietro lungo le $x$, se i = 2 allora va in avanti di uno lungo le $x$, etc. etc.). L'ampiezza di ciascun passo corrisponde a $a=1$.
- ***Random walk* continuo**: vengono generati casualmente gli angoli $\theta \in [0,\pi]$ e $\phi \in [0,2\pi]$ delle coordinate sferiche con cui si ricava il passo da eseguire in coordinate cartesiane da sommare alla posizione precedente. L'ampiezza di ciascun passo è fissato a $a=1$.

Alla fine di ciascun *random walk* (per entrambi i casi) ho valutato la distanza cartesiana dall'origine al variare del numero di passi $N\in [0,10^2]$, $\vec{r}_N = a \sqrt{x_N^2+y:N^2+z_N^2}$, ed il suo quadrato $|\vec{r}_N | ^2$.

La simulazione che ho lanciato prevede $M = 10^4$ realizzazioni di $|\vec{r}_N | ^2$ (nel caso discreto e continuo) sulle quali ho applicato il metodo del *data blocking* con $100$ blocchi. Ho iterato questa procedura facendo variare il numero di passi del *random walk* $N_{passi}=1,2,3,...,100$.

Si riporta nella cella sottostante i grafici $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ al variare del numero di passi $N_{passi}$ (per ottenere questi grafici ho utilizzato i valori medi e gli errori ottenuti usando $100$ blocchi per ogni passo).

Infine ho "fittato" i risultati ottenuti con la curva $y(N_{passi}) = a\sqrt{N_{passi}}$ (dove $a =1$ nella mia simulazione). Si può osservare come i dati della mia simulazione e la curva di fit presentino una buona compatibilità (tale affermazione è basata su una approssimativa analisi di lettura del grafico). Ciò indica che la simulazione da me eseguita, sia per il caso continuo che per quello discreto, è in accordo con la teoria matematica *random walk*; infatti esso è un processo diffusivo dove la deviazione standard è uguale a $a\sqrt{N_{passi}}$.

```python
plt.figure(figsize=(16,6))

plt.subplot(1,2,1)
N, discreto, error = np.loadtxt('Risultati/Exercise_02.2/Risultati02.2_discreto.out', usecols=(0,1,2), delimiter=' ', unpack='true')
x = np.linspace(0,100,100)
plt.errorbar(N,discreto,yerr=error, alpha=0.7,label='Dati')
plt.plot(x,np.sqrt(x), label='Fit')
plt.title("Random walk discreto su reticolo cubico con $a=1$")
plt.xlabel(r'$N_{passi}$')
plt.ylabel(r'$\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$')
plt.grid(True)
plt.legend()

plt.subplot(1,2,2)
N, continuo, error = np.loadtxt('Risultati/Exercise_02.2/Risultati02.2_continuo.out', usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(N,continuo,yerr=error, alpha=0.7,label='Dati')
plt.plot(x,np.sqrt(x),label='Fit')
plt.title("Randomwalk continuo su spazio tridimensionale con passo $a=1$")
plt.xlabel(r'$N_{passi}$')
plt.ylabel(r'$\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$')
plt.grid(True)
plt.legend()
plt.show()
```
